```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
Prop_data = pd.read_csv("all_props.csv")
Prop_data.head()
```

```python
Prop_data.columns
```

**csv 파일에서 프롭의 속도(V), Diameter, Pitch, RPM, 추력(Thrust)을 추출한 후 리스트 생성**

프롭의 속도(V)

```python
prop_v_list = []
for i in Prop_data["V"] :
  prop_v_list.append(i)
print(prop_v_list)
```

프롭의 직경(diameter)

```python
prop_diameter_list = []
for i in Prop_data["diameter"] :
  prop_diameter_list.append(i)
print(prop_diameter_list)
```

프롭의 Pitch

```python
prop_pitch_list = []
for i in Prop_data["pitch"] :
  prop_pitch_list.append(i)
print(prop_pitch_list)
```

프롭의 RPM

```python
prop_rpm_list = []
for i in Prop_data["rpm"] :
  prop_rpm_list.append(i)
print(prop_rpm_list)
```

프롭의 추력(Thrust)

```python
prop_thrust_list = []
for i in Prop_data["Thrust"] :
  prop_thrust_list.append(i)
print(prop_thrust_list)
```

검증 데이터 추가

```python
prop_check_list = []
prop_check_list.extend([[14.3,10.5,4.5,3000]])
prop_check_list.extend([[27.6,5.5,4.5,37000]])
prop_check_list.extend([[54.1,8.5,7,13000]])
prop_check_list.extend([[20.1,11,7,6000]])
print(prop_check_list)
```

**넘파이 배열로 전환**

numpy를 활용하여 프롭의 속도(V), Diameter, Pitch, RPM 데이터 통합

```python
prop_full = np.column_stack((prop_v_list, prop_diameter_list, prop_pitch_list, prop_rpm_list))
print(prop_full)
```

```python
prop_thrust = np.array(prop_thrust_list)
print(prop_thrust)
```

```python
prop_check = np.array(prop_check_list)
print(prop_check)
```

**사이킷런으로 훈련 세트 / 테스트 세트 나누기**

```python
from sklearn.model_selection import train_test_split
train_input, test_input, train_target, test_target = train_test_split(
    prop_full, prop_thrust, random_state=42)
```

**사이킷런의 변환기 - PolynomialFeatures**

```python
from sklearn.preprocessing import PolynomialFeatures
poly = PolynomialFeatures(include_bias=False)
poly.fit(train_input)
train_poly = poly.transform(train_input)
print(train_poly.shape)
```

```python
poly.get_feature_names_out()
```



> 14개의 특성이 어떤 조합으로 구성되어 있는지 출력



```python
test_poly = poly.transform(test_input)
```

**다중 회귀 모델 훈련**

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(train_poly, train_target)
```

훈련 세트의 점수

```python
print(lr.score(train_poly, train_target))
```

테스트 세트의 점수

```python
print(lr.score(test_poly, test_target))
```

**검증 데이터 예측값**

```python
check_poly = poly.transform(prop_check)
print(lr.predict(check_poly))
print(lr.predict(test_poly))
```

**특성을 늘려 5제곱까지 특성을 만들어 출력해보기**

```python
poly = PolynomialFeatures(degree=5, include_bias=False)
poly.fit(train_input)
train_poly = poly.transform(train_input)
test_poly = poly.transform(test_input)
check_poly = poly.transform(prop_check)
print(train_poly.shape)
```

```python
lr.fit(train_poly, train_target)
print(lr.score(train_poly, train_target))
```

```python
print(lr.score(test_poly, test_target))
```



> 입력된 데이터의 양이 크기 때문에 특성을 5제곱까지 늘려서 만들면 훈련 세트의 점수와 테스트 세트의 점수가 오른 것을 확인할 수 있다.



**규제**

사이킷런의 변환기 - StandardScaler

```python
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
ss.fit(train_poly)
train_scaled = ss.transform(train_poly)
test_scaled = ss.transform(test_poly)
check_scaled = ss.transform(check_poly)
```

**릿지 회귀**


> 계수의 제곱값을 기준으로 규제 적용



훈련 세트의 점수

```python
from sklearn.linear_model import Ridge
ridge = Ridge()
ridge.fit(train_scaled, train_target)
print(ridge.score(train_scaled, train_target))
```

테스트 세트의 점수

```python
print(ridge.score(test_scaled, test_target))
```

현재 확인된 정보에서는 훈련 세트와 테스트 세트의 정보가 과소적합, 과대적합이 모두 발생하지 않는 적절한 점수를 보임을 확인할 수 있다. **그러나 alpha의 값을 변경해보며 더 이상적인 값이 있는지 확인해보겠다.**

```python
train_score = []
test_score = []
alpha_list = [0.0000001, 0.0000001 ,0.00001, 0.0001 ,0.001, 0.01, 0.1, 1, 10, 100]
for alpha in alpha_list :
  ridge = Ridge(alpha=alpha)
  ridge.fit(train_scaled, train_target)
  train_score.append(ridge.score(train_scaled, train_target))
  test_score.append(ridge.score(test_scaled, test_target))
```

ridge.fit(train.scaled, train_target)



> 릿지 모델 훈련

train_score.append(ridge.score(train_scaled, train_target))

> 훈련 점수 저장

test_score.append(ridge.score(test_scaled, test_target))

> 테스트 점수 저장


```python
plt.plot(np.log10(alpha_list), train_score)
plt.plot(np.log10(alpha_list), test_score)
plt.xlabel('alpha')
plt.ylabel('R^2')
plt.show()
```

위의 그래프 형태를 보면 alpha값이 -5 이하로 가면 거의 수평적인 형태를 보이기 때문에 alpha값을 -5, 즉 10^-5=0.00001이다


> **alpha=0.00001**



```python
ridge = Ridge(alpha=0.00001)
ridge.fit(train_scaled, train_target)
print(ridge.score(train_scaled, train_target))
print(ridge.score(test_scaled, test_target))
```

위의 초기 점수는 0.9605, 0.9585였던 것에 비해 점수가 오른 것을 확인할 수 있다.

**검증 데이터 예측값**

```python
print(ridge.predict(check_scaled))
print(ridge.predict(test_scaled))
```

**라쏘 회귀**


> 계수를 절댓값을 기준으로 규제 적용



훈련 세트의 점수

```python
from sklearn.linear_model import Lasso
lasso = Lasso()
lasso.fit(train_scaled, train_target)
print(lasso.score(train_scaled, train_target))
```

테스트 세트의 점수

```python
print(lasso.score(test_scaled, test_target))
```

과소적합이 발생하는 것을 확인할 수 있다. 이는 규제가 너무 강했다는 것을 의미하므로 **alpha의 값을 최적의 값으로 수정해야한다.**

```python
train_score = []
test_score = []
alpha_list = [0.001, 0.01, 0.1, 1, 10, 100]
for alpha in alpha_list :
  lasso = Lasso(alpha=alpha, max_iter=10000)
  lasso.fit(train_scaled, train_target)
  train_score.append(lasso.score(train_scaled, train_target))
  test_score.append(lasso.score(test_scaled, test_target))
```

lasso.fit(train_scaled, train_target)



> 라쏘 모델 훈련

train_score.append(lasso.score(train_scaled, train_target))

> 훈련 점수 저장

test_score.append(lasso.score(test_scaled, test_target))

> 테스트 점수 저장


```python
plt.plot(np.log10(alpha_list), train_score)
plt.plot(np.log10(alpha_list), test_score)
plt.xlabel('alpha')
plt.ylabel('R^2')
plt.show()
```

위의 그래프에서는 훈련 세트의 그래프만이 표현된 것처럼 보이기 때문에 아래에 테스트 세트의 그래프를 따로 그리려고 한다.

```python
plt.plot(np.log10(alpha_list), test_score)
plt.xlabel('alpha')
plt.ylabel('R^2')
plt.show()
```

두 그래프는 동일하고 높아지는 경사가 완만해지는 경우가 -2, 즉 10^-2=0.01이다.

> **alpha = .0.01**



```python
lasso = Lasso(alpha=0.01)
lasso.fit(train_scaled, train_target)
print(lasso.score(train_scaled, train_target))
print(lasso.score(test_scaled, test_target))
```

라쏘 회귀 모델의 초기에는 과소적합이 발생했지만 alpha값을 수정한 후에는 적절한 점수값이 나오는 것을 확인할 수 있다.

**검증 데이터 예측값**

```python
print(lasso.predict(check_scaled))
print(lasso.predict(test_scaled))
```
