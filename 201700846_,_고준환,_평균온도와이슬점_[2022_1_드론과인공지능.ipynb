```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
Weather_df= pd.read_csv("TLS-weather-data-2013.csv")
Weather_df.head()
```

```python
Weather_df.columns
```

**csv 파일에서 평균온도(Mean TemperatureC)값과 이슬점(Dew PointC)값을 추출한 후 리스트 생성**

```python
mean_temp_list = []
for i in Weather_df["Mean TemperatureC"] :
  mean_temp_list.append(i)
print(mean_temp_list)
```

```python
dew_point_list = []
for i in Weather_df["Dew PointC"] :
  dew_point_list.append(i)
print(dew_point_list)
```

**넘파이 배열로 전환**

```python
mean_temp = np.array(mean_temp_list)
dew_point = np.array(dew_point_list)
```

```python
plt.scatter(mean_temp, dew_point)
plt.xlabel('Mean TemperatureC')
plt.ylabel('Dew PointC')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
train_input, test_input, train_target, test_target = train_test_split(mean_temp, dew_point, random_state=42)
```

```python
train_input = train_input.reshape(-1, 1)
test_input = test_input.reshape(-1, 1)
```

**k-최근접 이웃 회귀 알고리즘**

```python
from sklearn.neighbors import KNeighborsRegressor
knr = KNeighborsRegressor()
knr.fit(train_input, train_target)
```

```python
print(knr.score(test_input, test_target))
```

```python
from sklearn.metrics import mean_absolute_error
test_prediction = knr.predict(test_input)
mae = mean_absolute_error(test_target, test_prediction)
print(mae)
```

```python
print(knr.score(train_input, train_target))
```

**이웃 n값을 변경했을 때 훈련/테스트 세트의 R^2 점수**

```python
knr.n_neighbors = 10
knr.fit(train_input, train_target)
print(knr.score(train_input, train_target))
print(knr.score(test_input, test_target))
```

```python
for i in range (1,100) :
  knr.n_neighbors = i
  knr.fit(train_input, train_target)
  print('n_neighbors =',i)
  print(knr.score(train_input, train_target),knr.score(test_input, test_target),'\n')
```

위의 결과값으로 보았을 때 **최적의 이웃 n값은 9**임을 알 수 있다.

이웃 n값이 9일 때
훈련 세트의 R^2 값 : 0.8837274151521348

테스트 세트의 R^2 값 : 0.8687190301936257

```python
knr = KNeighborsRegressor()

x = np.arange(5, 35).reshape(-1, 1)

for n in [1,2,5,9,10,50,100,200]:
  knr.n_neighbors = n
  knr.fit(train_input, train_target)
  prediction = knr.predict(x)

  plt.scatter(train_input, train_target)
  plt.plot(x, prediction)
  plt.title('n_neighbors = {}'.format(n))
  plt.xlabel('Mean TemperatureC')
  plt.ylabel('Dew PointC')
  plt.show()
```

k-최근접 이웃 회귀 알고리즘에서 최적의 이웃 n값을 적용하여 훈련/테스트 세트의 R^2값을 구했지만 과소적합이 발생함을 알 수 있다.
따라서 선형 회귀 알고리즘을 이용해보고자 한다.

**선형 회귀**

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(train_input, train_target)
print(lr.coef_, lr.intercept_)
```

```python
point = np.arange(-5, 35)
plt.scatter(train_input, train_target)
plt.plot(point, lr.coef_*point + lr.intercept_)
plt.xlabel('Mean TemperatureC')
plt.ylabel('Dew PointC')
plt.show()
```
